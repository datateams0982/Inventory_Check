```python
def VWAP(row):  
    if (row['vol'] == 0) or (row['total'] == 0):
        return np.nan
    else:
        vwap = row['total']/row['vol']
        
        return vwap
    
def stock_query(end_date):
    
    start_date = (end_date - timedelta(days=150)).strftime('%Y%m%d')
    year = end_date.year
    month = end_date.month
    if month == 1:
        last_year = year - 1
        last_month = 12
    else:
        last_year = year
        last_month = month - 1
    last_month_start = date(last_year, last_month, 1).strftime('%Y%m%d')
    last_month_end = date(last_year, last_month, calendar.monthrange(last_year, last_month)[1]).strftime('%Y%m%d')
    end_date = end_date.strftime('%Y%m%d')

    stock_subquery = f'''SELECT d.[DATE] AS ts,
                            d.[STOCK_ID] AS StockNo,
                            TRY_CAST(d.[OPEN] AS DECIMAL(17, 2)) AS [open], 
                            TRY_CAST(d.[HIGH] AS DECIMAL(17, 2)) AS [high],
                            TRY_CAST(d.[LOW] AS DECIMAL(17, 2)) AS [low],
                            TRY_CAST(d.[CLOSE] AS DECIMAL(17, 2)) AS [close],
                            TRY_CAST(d.[VOLUME] AS DECIMAL(17, 2)) AS vol,
                            TRY_CAST(d.[AMOUNT] AS DECIMAL(17, 2)) AS total,
                            TRY_CAST(d.[CAPITAL] AS DECIMAL(17, 2)) AS capital,
                            TRY_CAST(e.[FOREIGN_VOL] AS DECIMAL(17, 2)) AS foreign_buy,
                            TRY_CAST(e.[INVEST_VOL] AS DECIMAL(17, 2)) AS investment_buy,
                            TRY_CAST(e.[DEALER_VOL] AS DECIMAL(17, 2)) AS dealer_buy, 
                            TRY_CAST(e.[FOREIGN_INV_RATIO] AS DECIMAL(17, 2)) AS foreign_ratio,
                            TRY_CAST(e.[INVEST_INV_RATIO] AS DECIMAL(17, 2)) AS investment_ratio,
                            TRY_CAST(e.[DEALER_INV_RATIO] AS DECIMAL(17, 2)) AS dealer_ratio,
                            TRY_CAST(e.[CORP_INV_RATIO] AS DECIMAL(17, 2)) AS corporation_ratio,
                            f.[INDUSTRY_ID] AS industry,
                            f.[EXCHANGE_DATE] AS [On_Date]
                    FROM 
                        (
                            SELECT *
                            FROM OpenData.dbo.CMONEY_DAILY_CLOSE
                            WHERE 
                                DATE <= {end_date} and DATE >= {start_date}
                                AND LEN(STOCK_ID) = 4
                        ) d 
                        LEFT JOIN 
                        (    SELECT * 
                            FROM OpenData.dbo.CMONEY_DAILY_CORP_TXN_SUMMARY
                            WHERE 
                                DATE <= {end_date} and DATE >= {start_date}
                                and LEN(STOCK_ID) = 4
                        ) e
                        ON d.STOCK_ID = e.STOCK_ID AND d.DATE = e.DATE
                        LEFT JOIN 
                        (    SELECT
                                STOCK_ID,
                                LISTING_TYPE,
                                INDUSTRY_ID,
                                EXCHANGE_DATE
                            FROM OpenData.dbo.CMONEY_LISTED_COMPANY_INFO
                            WHERE 
                                YEAR = {str(year)}
                        ) f
                        ON d.STOCK_ID = f.STOCK_ID
                        WHERE 
                            EXISTS(
                                SELECT *
                                FROM(
                                    SELECT 
                                        STOCK_ID, 
                                        MONTHLY_AMT,
                                        PERCENT_RANK() OVER(ORDER BY MONTHLY_AMT) AS PctRank                    

                                    FROM(
                                        SELECT 
                                            a.STOCK_ID, 
                                            SUM(TRY_CAST(AMOUNT AS DECIMAL(17,2))) as MONTHLY_AMT
                                        FROM OpenData.dbo.CMONEY_DAILY_CLOSE  a
                                        LEFT JOIN 
                                        (    SELECT
                                                STOCK_ID,
                                                LISTING_TYPE,
                                                INDUSTRY_ID
                                            FROM OpenData.dbo.CMONEY_LISTED_COMPANY_INFO
                                            WHERE 
                                                YEAR = {str(year)}
                                        ) b
                                        ON a.STOCK_ID = b.STOCK_ID
                                        LEFT JOIN 
                                        (    SELECT
                                                STOCK_ID,
                                                TERMINATE_DATE
                                            FROM OpenData.dbo.CMONEY_DELISTED_COMPANY_INFO
                                            WHERE 
                                                YEAR = {str(year)}
                                        ) c
                                        ON a.STOCK_ID = c.STOCK_ID
                                        WHERE 
                                            a.DATE between {last_month_start} and {last_month_end}
                                            and LEN(a.STOCK_ID) = 4
                                            and LEFT(a.STOCK_ID, 1) between '1' and '9'
                                            and LISTING_TYPE = 1
                                            and TERMINATE_DATE IS NULL
                                            and INDUSTRY_ID != '91'
                                        GROUP BY a.STOCK_ID
                                    ) p_rank
                                ) p_rank_filtered
                                where PctRank >= 0.4
                                and STOCK_ID = d.STOCK_ID
                            )
                        '''
    stock_df = send_query(stock_subquery, timecost = False, showcol = False, showlen = False)
    stock_df['VWAP'] = stock_df.apply(VWAP, axis=1)
    stock_df['VWAP'] = round(stock_df['VWAP'].astype(np.float64), 4)
    stock_df['ts'] = stock_df['ts'].apply(lambda x: datetime.strptime(str(x)[:4] + '-' + str(x)[4:6] + '-' + str(x)[6:], '%Y-%m-%d').date())
    stock_df['On_Date'] = stock_df['On_Date'].apply(lambda x: datetime.strptime(str(x)[:4] + '-' + str(x)[4:6] + '-' + str(x)[6:], '%Y-%m-%d').date())
    
    index_subquery = f'''SELECT [DATE] AS ts,
                                [STOCK_ID] AS StockNo,
                                TRY_CAST([OPEN] AS DECIMAL(17, 2)) AS [index_open], 
                                TRY_CAST([HIGH] AS DECIMAL(17, 2)) AS [index_high],
                                TRY_CAST([LOW] AS DECIMAL(17, 2)) AS [index_low],
                                TRY_CAST([CLOSE] AS DECIMAL(17, 2)) AS [index_close],
                                TRY_CAST([VOLUME] AS DECIMAL(17, 2)) AS index_vol
                        FROM OpenData.dbo.CMONEY_DAILY_CLOSE
                        WHERE 
                            STOCK_ID = 'TWA00' 
                            AND DATE <= {end_date} and DATE >= {start_date}
                        '''
    
    industry_subquery = f'''SELECT [DATE] AS ts,
                                [STOCK_ID] AS StockNo,
                                TRY_CAST([OPEN] AS DECIMAL(17, 2)) AS [index_open], 
                                TRY_CAST([HIGH] AS DECIMAL(17, 2)) AS [index_high],
                                TRY_CAST([LOW] AS DECIMAL(17, 2)) AS [index_low],
                                TRY_CAST([CLOSE] AS DECIMAL(17, 2)) AS [index_close],
                                TRY_CAST([VOLUME] AS DECIMAL(17, 2)) AS index_vol
                        FROM OpenData.dbo.CMONEY_DAILY_CLOSE
                        WHERE 
                            SUBSTRING(STOCK_ID,1,3) = 'TWB' 
                            AND DATE <= {end_date} and DATE >= {start_date}
                        '''
    
    eliminate_subquery = f'''SELECT [YEAR] AS year,
                                [STOCK_ID] AS StockNo,
                                [START_TRADE_DATE] AS [Restart_date], 
                                [TERMINAL_TRADE_DATE] AS [Stop_date]
                        FROM OpenData.dbo.CMONEY_REDUCT_SUMMARY
                        WHERE YEAR = {str(year)}
                        '''
    index_df = send_query(index_subquery, timecost = False, showcol = False, showlen = False)
    index_df['ts'] = index_df['ts'].apply(lambda x: datetime.strptime(str(x)[:4] + '-' + str(x)[4:6] + '-' + str(x)[6:], '%Y-%m-%d').date())
    industry_df = send_query(industry_subquery, timecost = False, showcol = False, showlen = False)
    industry_df['ts'] = industry_df['ts'].apply(lambda x: datetime.strptime(str(x)[:4] + '-' + str(x)[4:6] + '-' + str(x)[6:], '%Y-%m-%d').date())
    eliminate = send_query(eliminate_subquery, timecost = False, showcol = False, showlen = False) 
    eliminate['Restart_date'] = eliminate['Restart_date'].apply(lambda x: datetime.strptime(str(x)[:4] + '-' + str(x)[4:6] + '-' + str(x)[6:], '%Y-%m-%d').date())
    eliminate['Stop_date'] = eliminate['Stop_date'].apply(lambda x: datetime.strptime(str(x)[:4] + '-' + str(x)[4:6] + '-' + str(x)[6:], '%Y-%m-%d').date())
    
    return [stock_df, index_df, industry_df, eliminate]
```
