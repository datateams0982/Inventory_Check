```python
import numpy as np 
import pandas as pd 
import datetime
from datetime import datetime, date
from tqdm import tqdm_notebook as tqdm
import os
import codecs

from multiprocessing import Pool
from functools import partial
import Fundamental_function as func
```

```python
file_path = 'D:\\庫存健診開發\\data\\上市櫃公司基本資料\\'
file_list = os.listdir(file_path)
df_list = []

if __name__ == '__main__':
    with Pool(processes=12) as pool:
        for i, x in enumerate(tqdm(pool.imap_unordered(partial(func.read_fundamental, file_path=file_path), file_list), total=len(file_list)), 1):
                df_list.append(x)
                
fundamental = pd.concat(df_list, axis=0)
fundamental = fundamental.sort_values(by='year')
fundamental
```

```python
file_path = 'D:\\庫存健診開發\\data\\減資狀況\\'
file_list = os.listdir(file_path)
df_list = []

if __name__ == '__main__':
    with Pool(processes=12) as pool:
        for i, x in enumerate(tqdm(pool.imap_unordered(partial(func.read_eliminate, file_path=file_path), file_list), total=len(file_list)), 1):
                df_list.append(x)
                
eliminate = pd.concat(df_list, axis=0)
eliminate = eliminate.sort_values(by='year')
eliminate
```

```python
file_path = 'D:\\庫存健診開發\\data\\現金增資彙總表\\'
file_list = os.listdir(file_path)
df_list = []

if __name__ == '__main__':
    with Pool(processes=12) as pool:
        for i, x in enumerate(tqdm(pool.imap_unordered(partial(func.read_increase, file_path=file_path), file_list), total=len(file_list)), 1):
                df_list.append(x)
                
increase = pd.concat(df_list, axis=0)
increase = increase.sort_values(by='year')
increase
```

```python
fundamental.to_csv('D:\\庫存健診開發\\data\\Feature\\Fundamental.csv', index=False)
eliminate.to_csv('D:\\庫存健診開發\\data\\Feature\\eliminate.csv', index=False)
increase.to_csv('D:\\庫存健診開發\\data\\Feature\\increase.csv', index=False)
```

```python
data = pd.read_csv('D:\\庫存健診開發\\data\\Stock\\TWSE_Stock.csv', converters={'ts': str, 'StockNo': str, 'StockName': str})
data['ts'] = pd.to_datetime(data['ts'])
data['year'] = data['ts'].dt.year
data
```

```python
df_list = [group[1] for group in data.groupby(data['year'])]
output_list = []

for i, d in enumerate(tqdm(df_list, total=len(df_list))):
    year = d['year'].iloc[0]
    if year < 2009:
        d['total_num'] = np.nan
    else:
        d = pd.merge(d, fundamental[fundamental['year'] == year][['StockNo', 'total_num']], on='StockNo', how='left')
        
    output_list.append(d)
    
df = pd.concat(output_list, axis=0)
df = df.sort_values(by='StockNo')
df
```

```python
df_list = [group[1] for group in df.groupby(df['year'])]
output_list = []

for i, d in enumerate(tqdm(df_list, total=len(df_list))):
    year = d['year'].iloc[0]
    e = eliminate[eliminate['year'] == year]
    stock_list = e['StockNo'].unique().tolist()
    d['eliminate_period'] = 0
    d_list = [d[~d['StockNo'].isin(stock_list)]]
    for stock in stock_list:
        d2 = d[d['StockNo'] == stock]
        if len(d2) == 0:
            continue
        else:
            stop = e[e['StockNo'] == stock]['Stop_date'].tolist()
            restart = e[e['StockNo'] == stock]['Restart_date'].tolist()
            for i in range(len(stop)):
                index = d2[d2['ts'].dt.date < restart[i]][d2['ts'].dt.date >= stop[i]].index.tolist()
                d2.loc[index, 'eliminate_period'] = 1
        d_list.append(d2)
    
    d = pd.concat(d_list, axis=0)     
    output_list.append(d)
    
df_eliminate = pd.concat(output_list, axis=0)
df_eliminate = df_eliminate.sort_values(by='StockNo')
df_eliminate
```

```python
df_eliminate.to_csv('D:\\庫存健診開發\\data\\Stock\\TWSE_Stock_fundamental.csv', index=False)
```
