```python
import pandas as pd
```

```python
df_list = []
for i in range(0,5):
    this_df = pd.read_csv(f'high_valuation/train_{i}.csv')
    df_list.append(this_df)
df = pd.concat(df_list)
```

```python
df['type'] = 'TRAIN'
```

```python
df.describe()
```

```python
df.info()
```

```python
df2_list = []
for i in range(0,5):
    this_df = pd.read_csv(f'high_valuation/val_{i}.csv')
    df2_list.append(this_df)
df2 = pd.concat(df2_list)
```

```python
df2['type'] = 'VALIDATE'
```

```python
df2.describe()
```

```python
df2.info()
```

```python
df3_list = []
for i in range(0,5):
    this_df = pd.read_csv(f'high_valuation/test_{i}.csv')
    df3_list.append(this_df)
df3 = pd.concat(df3_list)
```

```python
df_test1 = pd.read_csv('D:/OneDrive - SinoPac/train_large.csv')
```

```python
prob_list = []
for i in range (len(df_test1.columns)):
    if '%' in df_test1.columns[i]:
        prob_list.append(df_test1.columns[i])
```

```python
fixed_list = [p.replace('%','') for p in prob_list]
```

```python
fixed_dict = dict(zip(prob_list,fixed_list))
```

```python
fixed_dict
```

```python
df_test1 = df_test1.rename(columns=fixed_dict)
```

```python
df_test1.to_csv('high_val_cluster_complete_2.csv', index=False)
```
