```python
data = pd.read_csv('D:\\庫存健診開發\\data\\Training\\Raw\\Cluster_7.csv', converters={'ts': str, 'StockNo': str, 'StockName': str})
data['ts'] = pd.to_datetime(data['ts'])
data
```

```python
from multiprocessing import Pool, TimeoutError, Lock, Value, current_process, cpu_count
from functools import partial
from tqdm import tqdm_notebook as tqdm
import datetime
from datetime import datetime, timedelta
```

```python
d = func.Normalization(data, scaler='minmax_one')
d
```

```python
df_list = [group[1] for group in data.groupby(data['StockNo'])]
train_list, test_list = [], []

if __name__ == '__main__':
    with Pool(processes=12) as pool:
        for i, x in enumerate(tqdm(pool.imap_unordered(partial(func.TrainTestSplit, SplitDate=datetime.strptime('2017-09-01', '%Y-%m-%d').date()), df_list), total=len(df_list)), 1):
            train_list.append(x[0])
            test_list.append(x[1])
```

```python
train_list
```

```python
df_list = [train_list, test_list]

X_train, Y_train, X_test, Y_test = [], [], [], []

if __name__ == '__main__':
    with Pool(processes=12) as pool:
        for i, stockdata in enumerate(df_list):
            if i == 0:
                for i, x in enumerate(tqdm(pool.imap_unordered(partial(func.create_dataset_classification, lookback=20, scaler='minmax_one'), stockdata), total=len(stockdata)), 1):
                    for xx in x[0]:
                        X_train.append(xx.tolist())
                    Y_train.extend(x[1])
            else:
                for i, x in enumerate(tqdm(pool.imap_unordered(partial(func.create_dataset_classification, lookback=20, scaler='minmax_one'), stockdata), total=len(stockdata)), 1):
                    for xx in x[0]:
                        X_test.append(xx.tolist())
                    Y_test.extend(x[1])
```

```python
(len(X_train) == len(Y_train)) and (len(X_test) == len(Y_test))
```

```python
Y_train, Y_test = func.categorical_transform(Y_train), func.categorical_transform(Y_test)
```

```python
save_list = [X_train, Y_train, X_test, Y_test]
```

```python
cluster = 7
import pickle
with open(f'D:\\庫存健診開發\\data\\Training\\processed\\Cluster_{cluster}_classification_minmax1', 'wb') as fp:
    pickle.dump(save_list, fp)
```
